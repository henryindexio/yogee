```python
import geopandas as gpd
from shapely.geometry import Point
import numpy as np
import pandas as pd
```

```python
import pickle

f = open('/home/henry/Insight/Yogee/Datasets/Model_dataset/ModelDf.pckl', 'rb')
ModelDf = pickle.load(f)
f.close()
```

```python
#Add columns for tranist dataset
#NanDfValues = np.zeros([np.shape(ModelDf)[0],2])
#NanDfValues[:] = np.nan
#NanDf = pd.DataFrame(NanDfValues,columns=['LongitudeBoundary','LatitudeBoundary'])

#ModelDf = pd.concat([ModelDf, NanDf], axis=1, sort=False)
ModelDf['LongitudeBoundary'] = pd.np.empty((np.shape(ModelDf)[0], 0)).tolist()
ModelDf['LatitudeBoundary'] = pd.np.empty((np.shape(ModelDf)[0], 0)).tolist()
ModelDf['LongitudeBoundary'] = ModelDf['LongitudeBoundary'].astype(object)
ModelDf['LongitudeBoundary'] = ModelDf['LongitudeBoundary'].astype(object)

```

```python
ZipcodeBoundaryDf = ModelDf[ModelDf['year']==2011]
ZipcodeBoundaryDf = ZipcodeBoundaryDf.reset_index(drop=True)
```

```python
# File path
ZipcodeShpLocation = "/home/henry/Downloads/Zipcode_boundary_shapefile/cb_2017_us_zcta510_500k.shp"

# Read the data
ZipcodeSh = gpd.read_file(ZipcodeShpLocation)
```

```python
#for i in range(0,ZipcodeBoundaryDf.shape[0]):
for i in range(0,ZipcodeBoundaryDf.shape[0]):
    ZipcodeStr = '{:05}'.format(int(ZipcodeBoundaryDf.loc[i,'zip']))
    ZipcodeShRow = ZipcodeSh[ZipcodeSh['ZCTA5CE10'] == ZipcodeStr]
    geometry = ZipcodeShRow['geometry'].iloc[0]
    polygon = []
    if str(type(geometry)) == "<class 'shapely.geometry.multipolygon.MultiPolygon'>":
        polygon = geometry[1]
    elif str(type(geometry)) == "<class 'shapely.geometry.polygon.Polygon'>":
        polygon = geometry
    LongitudeBoundary,LatitudeBoundary = polygon.exterior.coords.xy
    LongitudeBoundary = LongitudeBoundary.tolist()
    LatitudeBoundary = LatitudeBoundary.tolist()
    ZipcodeBoundaryDf.at[i,'LongitudeBoundary'] = LongitudeBoundary
    ZipcodeBoundaryDf.at[i,'LatitudeBoundary'] = LatitudeBoundary
```

```python
import pickle

f = open('/home/henry/Insight/Yogee/Datasets/Zipcode_Boundary_dataset/ZipcodeBoundaryDf.pckl', 'wb')
pickle.dump(ZipcodeBoundaryDf, f)
f.close()
```
