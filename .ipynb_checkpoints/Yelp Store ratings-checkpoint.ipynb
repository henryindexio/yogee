```python
import pickle

#Load Yoga model dataset
f = open('/home/henry/Insight/Yogee/Datasets/Model_dataset/ModelDf.pckl', 'rb')
ModelDf = pickle.load(f)
f.close()
```

```python
import pickle

#Load Yoga model dataset
f = open('/home/henry/Insight/Yogee/Datasets/Yelp_NY_Yoga_Studios_dataset/YogaDf.pckl', 'rb')
YogaDf = pickle.load(f)
f.close()
```

```python
#Initialize start year columns to yoga dataframe
NanDfValues = np.zeros([np.shape(YogaDf)[0],4])
NanDfValues[:] = np.nan
NanDf = pd.DataFrame(NanDfValues,columns=['startyear','PredictNewStudio','MeanRating','ReviewCount'])
YogaDf = pd.concat([YogaDf, NanDf], axis=1, sort=False)
```

```python
for i in range(0,YogaDf.shape[0]):
    dos = YogaDf.loc[i,'initial_dos_filing_date']    
    if isinstance(dos,str):
        dos = dos.replace('T00:00:00.000','')
        startyear = np.int64(dos[0:4])
        YogaDf.loc[i,'startyear'] = startyear
```

```python
YogaDf2015_6 = YogaDf[(YogaDf['startyear'] == 2016) | (YogaDf['startyear'] == 2015)]
YogaDf2015_6 = YogaDf2015_6.reset_index(drop=True)
```

```python
#Load Yelp API key and connect to API

f = open('/home/henry/Insight/APIKey/YelpAPIKey.pckl', 'rb')
YelpAPIKey = pickle.load(f)
f.close()
yelpapi = YelpAPI(YelpAPIKey, timeout_s=3.0)
```

```python
for i in range(29, YogaDf2015_6.shape[0]):
    YogaDf2015_6Row = YogaDf2015_6.iloc[i]
    Location = YogaDf2015_6Row['location']
    ZipcodeYoga = Location['zip_code']
    BusinessidYoga = YogaDf2015_6Row['id']
    PredictDfRow = PredictDf[PredictDf['zip'] == np.int(ZipcodeYoga)]
    ScorePredict = PredictDfRow['Score'].iloc[0]
    PredictNewStudio = (ScorePredict>5)*1
    ModelDfRow = ModelDf2015_6[ModelDf2015_6['zip'] == np.int(ZipcodeYoga)]
    ActualNewStudio = ModelDfRow['NewStudioNextYearBin'].iloc[0]
    YogaDf2015_6.loc[i,'PredictNewStudio'] = PredictNewStudio
    SearchResults = yelpapi.business_query(id=BusinessidYoga)
    YogaDf2015_6.loc[i,'MeanRating'] = SearchResults['rating']
    YogaDf2015_6.loc[i,'ReviewCount'] = SearchResults['review_count']
```

```python
from scipy import stats
Predict0 = YogaDf2015_6[YogaDf2015_6['PredictNewStudio']==0]['ReviewCount']
Predict1 = YogaDf2015_6[YogaDf2015_6['PredictNewStudio']==1]['ReviewCount']
stats.ks_2samp(Predict0,Predict1)
Predict0
```
